# 3. Feature engineering (Count vectorizer)

```python
#For data preprocessing
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

#For model creation and training
from sklearn.naive_bayes import MultinomialNB

#For model evaluation
from sklearn.metrics import classification_report, confusion_matrix
```

```python
vectorizer = CountVectorizer()
spamham_countVector = vectorizer.fit_transform(df['Message_body'])
```

```python
spamham_countVector.shape
```

```python
spamham_countVector
```

```python
X_train, X_test, y_train, y_test = train_test_split(spamham_countVector, y, test_size = 0.25)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## 4. train model

```python
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train,y_train)
```

## 5. evaluate model

```python
y_pred = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(metrics.accuracy_score(y_test, y_pred))
```
